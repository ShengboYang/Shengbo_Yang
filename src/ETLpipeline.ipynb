```python
df_laptimes = spark.read.csv("/Volumes/gr5069/raw/f1_data/lap_times.csv",header= True)
```

```python
display(df_laptimes)
```

```python
df_drivers= spark.read.csv("/Volumes/gr5069/raw/f1_data/drivers.csv",header= True)
```

```python
display(df_drivers)
```

```python
import pandas as pd
from pyspark.sql.functions import col, current_date, datediff, floor
df_driver = df_drivers.withColumn("age", floor(datediff(current_date(), col("dob")) / 365))

display(df_driver[["forename", "surname", "age"]])

```

## Transform Data for Drivers and added 'age'

```python
df_lap_drivers = df_driver.select('driverId','driverRef','number','code','forename','surname','nationality','age').join(df_laptimes, on=['driverId'])
```

```python
display(df_lap_drivers)
```

```python
df_lap_drivers.count()
```

```python
df_laptimes.count()
```

```python

df_lap_drivers = df_lap_drivers.groupby('age').agg({'milliseconds': 'avg'})
```

```python
display(df_lap_drivers)
```
